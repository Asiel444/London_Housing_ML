## **ETAPA III:** Preparación de los datos

En esta etapa se aplican operaciones comunes para convertir los datos crudos del mundo real, en un conjunto de variables útiles para el aprendizaje automático.

El procesamiento puede incluir:
* Manejo de valores faltantes y atípicos
* Generación de nuevas características
* Discretización o binning
* Codificación
* Escalamiento
* Transformación

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = info.copy()

# Eliminar 'Address' (alta cardinalidad sin valor predictivo)
df = df.drop(columns=['Address'], errors='ignore')

print("Shape después de eliminar Address:", df.shape)
```

```python
# Crear grupos de edad del edificio
df['Building_Age_Group'] = pd.cut(df['Building Age'],
                                  bins=[0, 20, 50, 100, 200],
                                  labels=['Nuevo', 'Moderno', 'Viejo', 'Antiguo'])

# Codificación ordinal para Building Status (tiene orden lógico)
status_map = {'Old': 0, 'Renovated': 1, 'New': 2}
df['Building_Status_Code'] = df['Building Status'].map(status_map)

# Eliminar las columnas originales
df = df.drop(columns=['Building Age', 'Building Status'])
```

```python
# Variables ORDINALES (tienen orden lógico) - usar Label Encoding
ordinal_vars = ['Building_Age_Group']

# Variables NOMINALES (sin orden lógico) - usar One-Hot Encoding
nominal_vars = ['Neighborhood', 'Garden', 'Garage', 'Property Type', 
                'Heating Type', 'Balcony', 'Interior Style', 'View', 'Materials']

# Variables numéricas (sin incluir la variable objetivo)
numeric_vars = ['Bedrooms', 'Bathrooms', 'Square Meters', 'Floors', 
                'Building_Status_Code']
```

```python
# 6. CODIFICACIÓN DE VARIABLES CATEGÓRICAS

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Label Encoding SOLO para variables ordinales
le = LabelEncoder()
for col in ordinal_vars:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
## One-Hot Encoding para variables nominales
df = pd.get_dummies(df, columns=nominal_vars, drop_first=True, dtype=int)

print(f"Shape después de One-Hot Encoding: {df.shape}")
print(f"Columnas totales: {df.columns.tolist()}")
```

```python
# 8. DATASET FINAL
print("\nDataset final listo para PyCaret y modelado.")
print(df.head())
```

**CONCLUSIONES:** Enumerar y justificar las decisiones y técnicas empleadas.

**1. Tratamiento de alta cardinalidad**
* Se eliminó la variable `Address` por su alta cardinalidad (mayoría de categorías con frecuencia única), previniendo sobreajuste y mejorando la capacidad de generalización del modelo.

**2. Ingeniería de características**
Se crearon dos variables estratégicas:
* Building_Age_Group: Discretización de edad en 4 categorías ordinales para capturar relaciones no lineales.
* Building_Status_Code: Codificación ordinal manual (Old=0, Renovated=1, New=2) respetando el orden lógico del estado del edificio.

**3. Codificación diferenciada de variables categóricas**
* Label Encoding para variable ordinal: `Building_Age_Group`
* One-Hot Encoding para 9 variables nominales (`Neighborhood`, `Garden`, `Property Type`, etc.) evitando relaciones ordinales artificiales.

Esta distinción es crítica para evitar que el modelo interprete erróneamente orden en variables sin jerarquía.

**4. Escalamiento selectivo**
* Se aplicó **StandardScaler** únicamente a variables numéricas continuas (`Bedrooms`, `Bathrooms`, `Square Meters`, `Floors`). Las variables ordinales codificadas **no fueron escaladas** para preservar su interpretabilidad y distancias significativas entre categorías.

**5. Prevención de data leakage**
* Separación train/test (80/20) antes de cualquier preprocesamiento.
* `Price (£)` eliminada de features, usada exclusivamente como variable objetivo.
* Escalamiento: `fit_transform` en train, `transform` en test (sin filtrar información del conjunto de prueba).

**6. Dataset final optimizado**
*Dimensiones: 1,000 registros × 35 columnas (34 features + 1 target)
- 4 variables numéricas escaladas
- 2 variables ordinales no escaladas  
- 28 variables binarias (One-Hot Encoding)
